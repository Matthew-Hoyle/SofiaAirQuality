```python
import pandas as pd
import numpy as np
import folium
import matplotlib.pyplot as plt
from matplotlib import colors
import seaborn as sns
import math
import json
```

```python
sofia_data = pd.read_csv("SofiaSensorDataCleaned.csv")
```

```python
sofia_data.head(10)
```

```python
sofia_grouped_all = sofia_data.groupby('district').mean().reset_index()
```

```python
sofia_grouped_all
```

```python
with open("Sofia_GeoMapping.json", "r") as j:
    SofiaGeoMap = json.loads(j.read())
```

```python
m = folium.Map(location = [42.69, 23.35], zoom_start = 11)
```

```python
cp = folium.Choropleth(
    geo_data=SofiaGeoMap,
    name="PM2.5",
    data=sofia_grouped_all,
    columns=["district", "PM2.5"],
    key_on="feature.id",
    fill_color="YlOrRd",
    fillnan_color = "white",
    line_color = 'black',
    fill_opacity=0.7,
    line_opacity=1,
    highlight = True,
    #tooltip = "District : {0}</i>".format("properties.name"),
    #legend_name="Unemployment Rate (%)",
).add_to(m)

folium.GeoJsonTooltip(['name'], aliases = ['District:']).add_to(cp.geojson)

folium.LayerControl().add_to(m)
```
