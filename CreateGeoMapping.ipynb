```python
import pandas as pd
import numpy as np
import folium
import matplotlib.pyplot as plt
from matplotlib import colors
import seaborn as sns
import json
```

```python
def create_mapping(coordinates, names):
    feature_list = []
    for i in range(len(coordinates)):
        dist_i = {'type' : 'Feature',
                  'id' : names[i],
                  'properties' : {'name' : names[i]},
                  'geometry' : {'type' : 'Polygon',
                                'coordinates' : [coordinates[i]]}}
        feature_list.append(dist_i)
    
    geojson_object = {'type' : 'FeatureCollection',
                      'features' : feature_list}
    
    return geojson_object
```

```python
with open("SofiaDistrictsRaw.kml", 'r') as f:
     SofiaDistrictsRaw = f.read()
```

```python
district_names = ["Sredets", "Vazrazhdane", "Oborishte", "Krasno selo", "Serdika", "Poduyane", "Slatina", "Izgrev", 
                  "Lozenets", "Triaditsa", "Krasna Polyana", "Ilinden", "Nadezhda", "Iskar", "Mladost", "Studentski",
                  "Lyulin", "Kremikovetsi", "Vitosha", "Ovcha kupel", "Vrabnits", "Novi Iskar", "Pancharevo", "Bankya"]
```

```python
intermed = [k.split(",0 ")[:-1] for k in[j.split("\n\t\t\t\t")[0] for j in [i.split('\t\t\t\t\t')[1] for i in SofiaDistrictsRaw.split("district")[2:]][1:]]]
intermed2 = []
for district in intermed:
    intermed2.append([[float(l.split(",")[0]), float(l.split(",")[1])] for l in district])
```

```python
intermed2
```

```python
district_mapping = create_mapping(intermed2, district_names)
```

```python
district_mapping
```

```python
m = folium.Map(location = [42.69, 23.35], zoom_start = 11)
```

```python
folium.Choropleth(
    geo_data=district_mapping,
    name="Sofia Districts",
    #data=districtsaq_df,
    #columns=["district", "PM10"],
    key_on="feature.id",
    #fill_color="YlGn",
    fill_color = 'gray',
    line_color = 'black',
    fill_opacity=0.3,
    line_opacity=1,
    highlight = True,
    #legend_name="Unemployment Rate (%)",
).add_to(m)

folium.LayerControl().add_to(m)
```
